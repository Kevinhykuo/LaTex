# Mapping by $\sin(z)$

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

from matplotlib import cycler
from IPython.display import Math, display

%config InlineBackend.figure_format = 'pdf'
```

## Mappping Vertical Line Segments

```python
def mapping(x, y):
    u = x**2 - y**2
    v = 2*x*y
    return u, v

def graph(X, Y, FigTitle1, FigTitle2, FileName):
    U, V = mapping(X, Y)
    
    n = np.shape(X)[0]
    color = plt.cm.Greys(np.linspace(0.1, 1, n))
    mp.rcParams['axes.prop_cycle'] = cycler('color', color)
    
    fig, ax = plt.subplots(1, 2, figsize = (10, 4))
    for i in range(0, n):
        # Make arrows on graphs
#         pos = int(len(X[i])) -1
#         ax[0].arrow(X[i, pos], Y[i, pos], X[i,pos]-X[i,pos-1],  Y[1,pos]-Y[1,pos-1], 
#                      shape='full', length_includes_head=True, width=0.1)
#         ax[1].arrow(U[i, pos], V[i,pos], U[i,pos]-U[i,pos-1],  V[1,pos]-V[1,pos-1], 
#                      shape='full', length_includes_head=True, width=0.1)
        
        ax[0].plot(X[i], Y[i])
        ax[1].plot(U[i], V[i])
        
        i += 1
    
    ax[0].set_title(FigTitle1)
    ax[1].set_title(FigTitle2)
    ax[0].set_xlabel("$x$")
    ax[0].set_ylabel("$y$")
    ax[1].set_xlabel("$u$")
    ax[1].set_ylabel("$v$")

#     # Move left y-axis and bottim x-axis to centre, passing through (0,0)
#     ax[0].spines['left'].set_position('center')
#     ax[0].spines['bottom'].set_position('center')
#     ax[1].spines['left'].set_position('center')
#     ax[1].spines['bottom'].set_position('center')

#     # Eliminate upper and right axes
#     ax[0].spines['right'].set_color('none')
#     ax[0].spines['top'].set_color('none')
#     ax[1].spines['right'].set_color('none')
#     ax[1].spines['top'].set_color('none')

#     # Show ticks in the left and lower axes only
#     ax[0].xaxis.set_ticks_position('bottom')
#     ax[0].yaxis.set_ticks_position('left')
#     ax[1].xaxis.set_ticks_position('bottom')
#     ax[1].yaxis.set_ticks_position('left')

#     ax[0].grid()
#     ax[1].grid()
    plt.tight_layout()
    plt.savefig(FileName, format='pdf',
        transparent=False, bbox_inches=None, pad_inches=0.1, metadata=None)
    plt.show()
```

```python
X, Y = np.mgrid[-1:1.2:0.2, 0:1+0.1:0.1]
graph(X, Y, "x = $c$", "$w = c^2 - y^2 + i2cy$", 'Mapping_by_z_Squared_V.pdf')
```

```python
Y2, X2 = np.mgrid[0:1+0.1:0.1, -1:1.05:0.05]
graph(X2, Y2, "y = $c$", "$w = x^2 - c^2 + i2xc$",'Mapping_by_z_Squared_H.pdf')
```
