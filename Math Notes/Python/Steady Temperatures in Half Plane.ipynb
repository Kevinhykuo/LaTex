# Steady Temperatures in a Half Plane

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

from matplotlib import cycler
from IPython.display import Math, display

%config InlineBackend.figure_format = 'pdf'
```

```python
def func(x, c):
    f = 1/(np.sin(np.pi * c))**2 - x**2 
    s = np.sqrt(f)
    P = s + 1/(np.tan(np.pi * c))
    N = -s + 1/(np.tan(np.pi * c))
    Y = np.append(P, N, axis=0)
    X = np.append(x, x, axis=0)
    return Y, X

def graph(x, c_start, c_end, c_step):
    c = np.arange(c_start, c_end, c_step)
    n = np.shape(c)[0]
    
    #color = plt.cm.Greys(np.arange(0.1, 1, n))
    #mp.rcParams['axes.prop_cycle'] = cycler('color', color)
    
    fig, ax = plt.subplots(1, 1, figsize = (4, 4))
    
    i = 0
    for i in range(0, n):
        Y, X = func(x, c[i])
        ax.plot(X, Y, 
                label='$c_1 = ${0:1.2f}'.format(c[i]), 
                c=(c[n-i-1], c[n-i-1], c[n-i-1]))
        i += 1
        
    ax.axis('equal')
    ax.set_xlim(-5, 10)
    ax.set_title('Isotherms $T(x,y) = c_1$ for $c_1 \in (0,1)$')
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.legend()
    plt.savefig('Steady_Temperatuers_In_Half_Plane.pdf', format='pdf',
        transparent=False, bbox_inches=None, pad_inches=0.1, metadata=None)
    plt.show()
```

```python
x = np.arange(-10, 10, 0.001)

graph(x, 0.1, 1, 0.1)
```
